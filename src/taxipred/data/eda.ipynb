```python
df["Time_of_Day"].value_counts().plot(kind="bar")
plt.title("Trips by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Count")
plt.show()
```

```python
df["Day_of_Week"].value_counts().plot(kind="bar")
plt.title("Trips by day of week")
plt.xlabel("Day of week")
plt.ylabel("Count")
plt.show()
```

```python
df["Weather"].value_counts().plot(kind="bar")
plt.title("Trips by weather")
plt.xlabel("weather")
plt.ylabel("Count")
plt.show()
```

```python
df["Traffic_Conditions"].value_counts().plot(kind="bar")
plt.title("Traffic conditions")
plt.xlabel("")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="Time_of_Day", y="Trip_Price", data=df)
plt.title("Trip Price per Time of Day")
plt.show()
```

```python
sns.scatterplot(x="Trip_Distance_km", y="Trip_Price", data=df)
plt.title("Relationship between Trip Distance and Trip Price")
plt.xlabel("Trip Distance (km)")
plt.ylabel("Trip Price")
plt.show()
```
